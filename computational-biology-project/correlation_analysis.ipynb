## Load chemical data and normalise/transform it

```python
# Load chemical data
chemical_data = pd.read_csv("water_chemicals.tsv", delimiter='\t')
```

```python
# Remove the 'CAS' column
chemical_data = chemical_data.drop(columns=['CAS'])
```

```python
print(chemical_data.columns)
chemical_data
```

```python
# Remove rows with all 0s

# Select only numeric columns
numeric_columns = chemical_data.drop('ChemName', axis=1)

# Convert numeric columns to numeric data type
numeric_columns = numeric_columns.apply(pd.to_numeric, errors='coerce')

# Remove rows with all values less than or equal to 0
chemical_data = chemical_data[numeric_columns.gt(0).any(axis=1)]

# Display the modified DataFrame
print(chemical_data)
```

```python
#Remove rows with less than 3 values
# Count non-zero values in each row
non_zero_counts = (chemical_data.iloc[:, 1:] > 0).sum(axis=1)

# Filter out rows with less than 3 non-zero values
chemical_data_fil = chemical_data[non_zero_counts >= 3]

chemical_data_fil.shape

print(chemical_data_fil)
```

```python
# Check for NaN values in the original data
nan_values = chemical_data_fil.isna().sum()
print(nan_values)
```

```python
# Extract the locations from the chemical data DataFrame
locations = chemical_data.columns[1:]  # Exclude 'ChemName'

# Create a new DataFrame to hold the repeated values
new_data = pd.DataFrame()

# Repeat the values for each location and suffix
for loc in locations:
    loc_values = chemical_data[loc].values
    repeated_values = pd.concat([pd.Series(loc_values)] * 6, axis=1)
    repeated_values.columns = [f'{loc}A{i+1}' for i in range(6)]
    new_data = pd.concat([new_data, repeated_values], axis=1)

# Repeat the 'ChemName' column 6 times
chem_name_repeated = pd.concat([chemical_data['ChemName']] * 6, ignore_index=True)

# Concatenate the 'ChemName' column with the new repeated data
chemical_data_repeated = pd.concat([chem_name_repeated, new_data], axis=1)

# Display the modified DataFrame
print(chemical_data_repeated)
```

```python
# Remove rows with all 0s

# Select only numeric columns
numeric_columns = chemical_data_repeated.drop('ChemName', axis=1)

# Convert numeric columns to numeric data type
numeric_columns = numeric_columns.apply(pd.to_numeric, errors='coerce')

# Remove rows with all values less than or equal to 0
chemical_data_repeated = chemical_data_repeated[numeric_columns.gt(0).any(axis=1)]

# Remove rows where the sum of values is 0
chemical_data_repeated = chemical_data_repeated[chemical_data_repeated.iloc[:, 1:].sum(axis=1) != 0]

# Display the modified DataFrame
print(chemical_data_repeated)
```

```python
#Remove rows with less than 3 values
# Count non-zero values in each row
non_zero_counts = (chemical_data_repeated.iloc[:, 1:] > 0).sum(axis=1)

# Filter out rows with less than 3 non-zero values
chemical_data_repeated = chemical_data_repeated[non_zero_counts >= 3]

chemical_data_repeated.shape

print(chemical_data_repeated)
```
